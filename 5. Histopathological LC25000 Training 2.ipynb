```python
import os
import time
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from itertools import cycle

# Metrics & Sklearn
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support, 
                             confusion_matrix, classification_report, roc_curve, auc, 
                             precision_recall_curve, average_precision_score, 
                             matthews_corrcoef, roc_auc_score, f1_score, precision_score, recall_score)
from sklearn.preprocessing import label_binarize
from sklearn.manifold import TSNE
from sklearn.calibration import calibration_curve

# PyTorch
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, ConcatDataset, Dataset
from torchvision import datasets, transforms, models
from torchsummary import summary

# XAI
from captum.attr import LayerGradCam
from captum.attr import visualization as viz

# ==================================================================================
# 1. CONFIGURATION
# ==================================================================================
BASE_PATH = '/kaggle/input/lung-and-colon-cancer-histopathological-images/lung_colon_image_set'

CONFIG = {
    'LUNG_DIR': os.path.join(BASE_PATH, 'lung_image_sets'),
    'COLON_DIR': os.path.join(BASE_PATH, 'colon_image_sets'),
    'IMG_SIZE': 224,
    'BATCH_SIZE': 32,
    'MAIN_EPOCHS': 20,       
    'LEARNING_RATE': 1e-4,
    'DEVICE': torch.device("cuda" if torch.cuda.is_available() else "cpu")
}

print(f"ðŸš€ Device: {CONFIG['DEVICE']}")

# ==================================================================================
# 2. MODEL ARCHITECTURE (Effi-DualAtt-KAN)
# ==================================================================================
class CoordinateAttention(nn.Module):
    def __init__(self, inp, oup, reduction=32):
        super(CoordinateAttention, self).__init__()
        self.pool_h = nn.AdaptiveAvgPool2d((None, 1))
        self.pool_w = nn.AdaptiveAvgPool2d((1, None))
        mip = max(8, inp // reduction)
        self.conv1 = nn.Conv2d(inp, mip, kernel_size=1, stride=1, padding=0)
        self.bn1 = nn.BatchNorm2d(mip)
        self.act = nn.Hardswish()
        self.conv_h = nn.Conv2d(mip, oup, kernel_size=1, stride=1, padding=0)
        self.conv_w = nn.Conv2d(mip, oup, kernel_size=1, stride=1, padding=0)

    def forward(self, x):
        identity = x
        n, c, h, w = x.size()
        x_h = self.pool_h(x)
        x_w = self.pool_w(x).permute(0, 1, 3, 2)
        y = torch.cat([x_h, x_w], dim=2)
        y = self.conv1(y)
        y = self.bn1(y)
        y = self.act(y)
        x_h, x_w = torch.split(y, [h, w], dim=2)
        x_w = x_w.permute(0, 1, 3, 2)
        a_h = torch.sigmoid(self.conv_h(x_h))
        a_w = torch.sigmoid(self.conv_w(x_w))
        return identity * a_w * a_h

class SimAM(torch.nn.Module):
    def __init__(self, e_lambda=1e-4):
        super(SimAM, self).__init__()
        self.activaton = nn.Sigmoid()
        self.e_lambda = e_lambda
    def forward(self, x):
        b, c, h, w = x.size()
        n = w * h - 1
        x_minus_mu_square = (x - x.mean(dim=[2,3], keepdim=True)).pow(2)
        y = x_minus_mu_square / (4 * (x_minus_mu_square.sum(dim=[2,3], keepdim=True) / n + self.e_lambda)) + 0.5
        return x * self.activaton(y)

class KANLinear(nn.Module):
    def __init__(self, in_features, out_features, grid_size=5):
        super(KANLinear, self).__init__()
        self.base_weight = nn.Parameter(torch.Tensor(out_features, in_features))
        self.spline_weight = nn.Parameter(torch.Tensor(out_features, in_features, grid_size))
        self.reset_parameters()
    def reset_parameters(self):
        nn.init.kaiming_uniform_(self.base_weight, a=5**0.5)
        nn.init.normal_(self.spline_weight, 0.0, 1.0)
    def forward(self, x):
        base_output = torch.nn.functional.linear(x, self.base_weight)
        spline_contribution = torch.sum(self.spline_weight.mean(dim=2) * x.unsqueeze(1), dim=2)
        return base_output + spline_contribution

class DualAtt_KAN_Model(nn.Module):
    def __init__(self, num_classes):
        super(DualAtt_KAN_Model, self).__init__()
        original_model = models.efficientnet_b0(weights='IMAGENET1K_V1')
        self.features = original_model.features
        self.coord_att = CoordinateAttention(inp=1280, oup=1280)
        self.sim_att = SimAM()
        self.avgpool = nn.AdaptiveAvgPool2d(1)
        self.dropout = nn.Dropout(p=0.3)
        self.kan_head = KANLinear(1280, num_classes)

    def forward(self, x):
        x = self.features(x)
        x = self.coord_att(x) 
        x = self.sim_att(x)   
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.dropout(x)
        x = self.kan_head(x)
        return x

# ==================================================================================
# 3. DATA PIPELINE
# ==================================================================================
def get_dataloaders():
    tfm = transforms.Compose([
        transforms.Resize((CONFIG['IMG_SIZE'], CONFIG['IMG_SIZE'])),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    if not os.path.exists(CONFIG['LUNG_DIR']): return None, None, None
    ds_lung = datasets.ImageFolder(CONFIG['LUNG_DIR'])
    ds_colon = datasets.ImageFolder(CONFIG['COLON_DIR'])
    offset = len(ds_lung.classes)
    ds_colon.targets = [t + offset for t in ds_colon.targets]
    ds_colon.samples = [(s[0], s[1] + offset) for s in ds_colon.samples]
    combined_classes = ds_lung.classes + ds_colon.classes
    full_dataset = ConcatDataset([ds_lung, ds_colon])
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_raw, val_raw = random_split(full_dataset, [train_size, val_size])
    
    class TransformSubset(Dataset):
        def __init__(self, subset, transform=None):
            self.subset = subset; self.transform = transform
        def __getitem__(self, index):
            x, y = self.subset[index]
            if self.transform: x = self.transform(x)
            return x, y
        def __len__(self): return len(self.subset)

    train_loader = DataLoader(TransformSubset(train_raw, tfm), batch_size=CONFIG['BATCH_SIZE'], shuffle=True, num_workers=2)
    val_loader = DataLoader(TransformSubset(val_raw, tfm), batch_size=CONFIG['BATCH_SIZE'], shuffle=False, num_workers=2)
    return train_loader, val_loader, combined_classes

# ==================================================================================
# 4. ENHANCED TRAINING LOOP (TRACKS ALL METRICS PER EPOCH)
# ==================================================================================
def train_engine(model, train_loader, val_loader, epochs, num_classes):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=CONFIG['LEARNING_RATE'])
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=2, factor=0.5, verbose=True)
    
    # Initialize History Dictionary
    history = {
        'train_loss': [], 'train_acc': [], 
        'val_loss': [], 'val_acc': [],
        'val_precision': [], 'val_recall': [], 'val_f1': [],
        'val_auc': [], 'val_mcc': []
    }
    
    best_acc = 0.0
    best_wts = copy.deepcopy(model.state_dict())
    
    print(f"\n>>> Starting Training for {epochs} Epochs...")
    
    for epoch in range(epochs):
        # --- TRAIN ---
        model.train()
        r_loss, r_corr = 0.0, 0
        
        for inputs, labels in tqdm(train_loader, desc=f"Ep {epoch+1}", leave=False):
            inputs, labels = inputs.to(CONFIG['DEVICE']), labels.to(CONFIG['DEVICE'])
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            _, preds = torch.max(outputs, 1)
            r_loss += loss.item() * inputs.size(0)
            r_corr += (preds == labels).sum().item()
            
        e_loss = r_loss / len(train_loader.dataset)
        e_acc = r_corr / len(train_loader.dataset)
        
        # --- VALIDATION (Full Metric Collection) ---
        model.eval()
        v_loss = 0.0
        all_preds = []
        all_labels = []
        all_probs = []
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(CONFIG['DEVICE']), labels.to(CONFIG['DEVICE'])
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                probs = torch.softmax(outputs, dim=1)
                _, preds = torch.max(outputs, 1)
                
                v_loss += loss.item() * inputs.size(0)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_probs.extend(probs.cpu().numpy())
                
        v_loss = v_loss / len(val_loader.dataset)
        
        # Calculate Epoch Metrics
        v_acc = accuracy_score(all_labels, all_preds)
        v_prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
        v_rec = recall_score(all_labels, all_preds, average='macro', zero_division=0)
        v_f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
        v_mcc = matthews_corrcoef(all_labels, all_preds)
        
        # AUC Calculation (Handle binary/multi-class)
        try:
            if num_classes == 2:
                v_auc = roc_auc_score(all_labels, np.array(all_probs)[:, 1])
            else:
                v_auc = roc_auc_score(all_labels, all_probs, multi_class='ovr')
        except:
            v_auc = 0.0

        scheduler.step(v_acc)
        
        # Store in History
        history['train_loss'].append(e_loss)
        history['train_acc'].append(e_acc)
        history['val_loss'].append(v_loss)
        history['val_acc'].append(v_acc)
        history['val_precision'].append(v_prec)
        history['val_recall'].append(v_rec)
        history['val_f1'].append(v_f1)
        history['val_auc'].append(v_auc)
        history['val_mcc'].append(v_mcc)
        
        print(f"Ep {epoch+1}: TrAcc={e_acc:.4f} | ValAcc={v_acc:.4f} | ValF1={v_f1:.4f} | ValAUC={v_auc:.4f}")
        
        if v_acc > best_acc:
            best_acc = v_acc
            best_wts = copy.deepcopy(model.state_dict())
            
    model.load_state_dict(best_wts)
    return model, history

# ==================================================================================
# 5. VISUALIZATION: TRAINING HISTORY PLOTS
# ==================================================================================
def plot_training_history(history):
    print("\n" + "="*40 + "\n   TRAINING METRICS PER EPOCH\n" + "="*40)
    epochs = range(1, len(history['train_loss']) + 1)
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. Loss & Accuracy
    axes[0,0].plot(epochs, history['train_loss'], label='Train Loss', color='blue', linestyle='--')
    axes[0,0].plot(epochs, history['val_loss'], label='Val Loss', color='red')
    axes[0,0].set_title("Loss Curve")
    axes[0,0].set_xlabel("Epochs")
    axes[0,0].legend()
    
    axes[0,1].plot(epochs, history['train_acc'], label='Train Acc', color='blue', linestyle='--')
    axes[0,1].plot(epochs, history['val_acc'], label='Val Acc', color='red')
    axes[0,1].set_title("Accuracy Curve")
    axes[0,1].set_xlabel("Epochs")
    axes[0,1].legend()
    
    # 2. Precision, Recall, F1
    axes[1,0].plot(epochs, history['val_precision'], label='Precision', color='purple')
    axes[1,0].plot(epochs, history['val_recall'], label='Recall', color='orange')
    axes[1,0].plot(epochs, history['val_f1'], label='F1-Score', color='green')
    axes[1,0].set_title("Validation: Precision, Recall, F1")
    axes[1,0].set_xlabel("Epochs")
    axes[1,0].legend()
    
    # 3. AUC & MCC
    axes[1,1].plot(epochs, history['val_auc'], label='ROC-AUC', color='magenta')
    axes[1,1].plot(epochs, history['val_mcc'], label='MCC', color='brown')
    axes[1,1].set_title("Validation: AUC & MCC")
    axes[1,1].set_xlabel("Epochs")
    axes[1,1].legend()
    
    plt.tight_layout()
    plt.show()

# ==================================================================================
# 6. SPECIFIC THRESHOLD VALIDATION (0.2, 0.4, 0.6, 0.8)
# ==================================================================================
def run_threshold_analysis(model, val_loader, classes):
    """
    Analyzes validation performance specifically at thresholds 0.2, 0.4, 0.6, 0.8
    """
    print("\n" + "="*50)
    print("   VALIDATION AT SPECIFIC THRESHOLDS (0.2, 0.4, 0.6, 0.8)")
    print("="*50)
    
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc="Inference"):
            inputs = inputs.to(CONFIG['DEVICE'])
            outputs = torch.softmax(model(inputs), dim=1)
            probs, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
            all_probs.extend(probs.cpu().numpy())
            
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)
    
    thresholds = [0.2, 0.4, 0.6, 0.8]
    results = []
    
    for t in thresholds:
        # Filter: Only keep predictions where confidence >= t
        mask = all_probs >= t
        
        if np.sum(mask) > 0:
            filt_preds = all_preds[mask]
            filt_labels = all_labels[mask]
            
            acc = accuracy_score(filt_labels, filt_preds)
            prec = precision_score(filt_labels, filt_preds, average='macro', zero_division=0)
            rec = recall_score(filt_labels, filt_preds, average='macro', zero_division=0)
            f1 = f1_score(filt_labels, filt_preds, average='macro', zero_division=0)
            kept = np.sum(mask)
            retention = kept / len(all_labels)
        else:
            acc, prec, rec, f1, retention, kept = 0, 0, 0, 0, 0, 0
            
        results.append({
            'Threshold': t,
            'Accuracy': acc,
            'Precision': prec,
            'Recall': rec,
            'F1-Score': f1,
            'Retention': retention,
            'Samples': kept
        })
    
    # Display Dataframe
    df = pd.DataFrame(results)
    print("\n--- Threshold Performance Table ---")
    print(df.to_string(index=False, formatters={
        'Accuracy': '{:.4f}'.format, 'Precision': '{:.4f}'.format, 
        'Recall': '{:.4f}'.format, 'F1-Score': '{:.4f}'.format, 'Retention': '{:.2%}'.format
    }))
    
    # Plot
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax1.plot(df['Threshold'], df['Accuracy'], marker='o', label='Accuracy', color='blue', linewidth=2)
    ax1.plot(df['Threshold'], df['F1-Score'], marker='^', label='F1-Score', color='green', linewidth=2)
    ax1.set_xlabel('Confidence Threshold')
    ax1.set_ylabel('Metric Score')
    ax1.set_ylim(0.8, 1.01)
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='upper left')
    
    ax2 = ax1.twinx()
    ax2.plot(df['Threshold'], df['Retention'], marker='s', label='Retention Rate', color='orange', linestyle='--')
    ax2.set_ylabel('Retention Rate')
    ax2.legend(loc='lower right')
    
    plt.title('Performance Metrics at Thresholds [0.2, 0.4, 0.6, 0.8]')
    plt.show()

# ==================================================================================
# 7. MAIN EXECUTION
# ==================================================================================
if __name__ == "__main__":
    # 1. Load
    train_loader, val_loader, class_names = get_dataloaders()
    
    if train_loader:
        # 2. Init
        model = DualAtt_KAN_Model(num_classes=len(class_names)).to(CONFIG['DEVICE'])
        print("\nModel Initialized. Total Params:", sum(p.numel() for p in model.parameters()))
        
        # 3. Train (With Full Metric Tracking)
        model, history = train_engine(model, train_loader, val_loader, epochs=CONFIG['MAIN_EPOCHS'], num_classes=len(class_names))
        
        # 4. Plot Training History (Acc, Loss, Prec, Rec, F1, AUC, MCC)
        plot_training_history(history)
        
        # 5. Run Specific Threshold Analysis (0.2, 0.4, 0.6, 0.8)
        run_threshold_analysis(model, val_loader, class_names)
```
