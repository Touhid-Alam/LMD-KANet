```python
import os
import time
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from itertools import cycle

# Sklearn Metrics & Tools
from sklearn.metrics import (roc_curve, auc, precision_recall_curve, average_precision_score, 
                             confusion_matrix, classification_report, matthews_corrcoef)
from sklearn.preprocessing import label_binarize
from sklearn.manifold import TSNE
from sklearn.calibration import calibration_curve

# PyTorch
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, ConcatDataset, Dataset
from torchvision import datasets, transforms, models

# XAI (Explainable AI)
from captum.attr import LayerGradCam
from captum.attr import visualization as viz

# ==================================================================================
# 1. CONFIGURATION & PATH VERIFICATION
# ==================================================================================
# Exact Root Path for Kaggle
BASE_PATH = '/kaggle/input/lung-and-colon-cancer-histopathological-images/lung_colon_image_set'

CONFIG = {
    'LUNG_DIR': os.path.join(BASE_PATH, 'lung_image_sets'),
    'COLON_DIR': os.path.join(BASE_PATH, 'colon_image_sets'),
    'IMG_SIZE': 224,
    'BATCH_SIZE': 32,
    'EPOCHS': 20,       # Adjust epochs as needed
    'LEARNING_RATE': 1e-4,
    'DEVICE': torch.device("cuda" if torch.cuda.is_available() else "cpu")
}

print(f"ðŸš€ Processing on Device: {CONFIG['DEVICE']}")

# Verify Paths
if os.path.exists(CONFIG['LUNG_DIR']) and os.path.exists(CONFIG['COLON_DIR']):
    print(f"âœ… Dataset Paths Verified.")
else:
    print("âŒ ERROR: Paths not found. Check directory structure.")

# ==================================================================================
# 2. MODEL ARCHITECTURE: EffiKAN-Attention
# ==================================================================================
class CoordinateAttention(nn.Module):
    def __init__(self, inp, oup, reduction=32):
        super(CoordinateAttention, self).__init__()
        self.pool_h = nn.AdaptiveAvgPool2d((None, 1))
        self.pool_w = nn.AdaptiveAvgPool2d((1, None))
        mip = max(8, inp // reduction)
        self.conv1 = nn.Conv2d(inp, mip, kernel_size=1, stride=1, padding=0)
        self.bn1 = nn.BatchNorm2d(mip)
        self.act = nn.Hardswish()
        self.conv_h = nn.Conv2d(mip, oup, kernel_size=1, stride=1, padding=0)
        self.conv_w = nn.Conv2d(mip, oup, kernel_size=1, stride=1, padding=0)

    def forward(self, x):
        identity = x
        n, c, h, w = x.size()
        x_h = self.pool_h(x)
        x_w = self.pool_w(x).permute(0, 1, 3, 2)
        y = torch.cat([x_h, x_w], dim=2)
        y = self.conv1(y)
        y = self.bn1(y)
        y = self.act(y)
        x_h, x_w = torch.split(y, [h, w], dim=2)
        x_w = x_w.permute(0, 1, 3, 2)
        a_h = torch.sigmoid(self.conv_h(x_h))
        a_w = torch.sigmoid(self.conv_w(x_w))
        out = identity * a_w * a_h
        return out

class KANLinear(nn.Module):
    def __init__(self, in_features, out_features, grid_size=5):
        super(KANLinear, self).__init__()
        self.base_weight = nn.Parameter(torch.Tensor(out_features, in_features))
        self.spline_weight = nn.Parameter(torch.Tensor(out_features, in_features, grid_size))
        self.reset_parameters()

    def reset_parameters(self):
        nn.init.kaiming_uniform_(self.base_weight, a=5**0.5)
        nn.init.normal_(self.spline_weight, 0.0, 1.0)

    def forward(self, x):
        base_output = torch.nn.functional.linear(x, self.base_weight)
        spline_contribution = torch.sum(
            self.spline_weight.mean(dim=2) * x.unsqueeze(1), dim=2
        )
        return base_output + spline_contribution

class EffiKAN_Attention_Model(nn.Module):
    def __init__(self, num_classes):
        super(EffiKAN_Attention_Model, self).__init__()
        # EfficientNet-B0 Backbone
        original_model = models.efficientnet_b0(weights='IMAGENET1K_V1')
        self.features = original_model.features
        # 1280 is output channel size of B0
        self.attention = CoordinateAttention(inp=1280, oup=1280)
        self.avgpool = nn.AdaptiveAvgPool2d(1)
        self.dropout = nn.Dropout(p=0.3)
        self.kan_head = KANLinear(1280, num_classes)

    def forward(self, x):
        x = self.features(x)
        x = self.attention(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.dropout(x)
        x = self.kan_head(x)
        return x

# ==================================================================================
# 3. DATA LOADING (MERGING & SPLITTING)
# ==================================================================================
def get_combined_dataloaders():
    # Augmentation
    train_tfm = transforms.Compose([
        transforms.Resize((CONFIG['IMG_SIZE'], CONFIG['IMG_SIZE'])),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(20),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    
    val_tfm = transforms.Compose([
        transforms.Resize((CONFIG['IMG_SIZE'], CONFIG['IMG_SIZE'])),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    print("\n--- Loading Datasets ---")
    ds_lung = datasets.ImageFolder(CONFIG['LUNG_DIR'])
    ds_colon = datasets.ImageFolder(CONFIG['COLON_DIR'])

    print(f"Lung Classes [0-2]: {ds_lung.classes}")
    print(f"Colon Classes [0-1]: {ds_colon.classes}")

    # SHIFT LOGIC: Move Colon indices to follow Lung indices
    offset = len(ds_lung.classes)
    ds_colon.targets = [t + offset for t in ds_colon.targets]
    ds_colon.samples = [(s[0], s[1] + offset) for s in ds_colon.samples]
    
    combined_classes = ds_lung.classes + ds_colon.classes
    print(f"Combined Classes: {combined_classes}")
    
    full_dataset = ConcatDataset([ds_lung, ds_colon])
    
    # Split
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_raw, val_raw = random_split(full_dataset, [train_size, val_size])
    
    # Wrapper to enforce transforms on subsets
    class TransformSubset(Dataset):
        def __init__(self, subset, transform=None):
            self.subset = subset
            self.transform = transform
        def __getitem__(self, index):
            x, y = self.subset[index]
            if self.transform:
                x = self.transform(x)
            return x, y
        def __len__(self):
            return len(self.subset)

    train_loader_set = TransformSubset(train_raw, train_tfm)
    val_loader_set = TransformSubset(val_raw, val_tfm)

    train_loader = DataLoader(train_loader_set, batch_size=CONFIG['BATCH_SIZE'], shuffle=True, num_workers=2)
    val_loader = DataLoader(val_loader_set, batch_size=CONFIG['BATCH_SIZE'], shuffle=False, num_workers=2)
    
    return train_loader, val_loader, combined_classes

# ==================================================================================
# 4. TRAINING LOOP
# ==================================================================================
def train_model(model, train_loader, val_loader, epochs):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=CONFIG['LEARNING_RATE'], weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5, verbose=True)
    
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())

    print(f"\n--- Starting Training for {epochs} Epochs ---")
    
    for epoch in range(epochs):
        start_time = time.time()
        model.train()
        running_loss, running_corrects = 0.0, 0
        
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(CONFIG['DEVICE']), labels.to(CONFIG['DEVICE'])
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            _, preds = torch.max(outputs, 1)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)
            pbar.set_postfix({'loss': loss.item()})
            
        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = running_corrects.double() / len(train_loader.dataset)
        
        # Validation
        model.eval()
        val_loss, val_corrects = 0.0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(CONFIG['DEVICE']), labels.to(CONFIG['DEVICE'])
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                _, preds = torch.max(outputs, 1)
                val_loss += loss.item() * inputs.size(0)
                val_corrects += torch.sum(preds == labels.data)
                
        val_epoch_loss = val_loss / len(val_loader.dataset)
        val_epoch_acc = val_corrects.double() / len(val_loader.dataset)
        
        scheduler.step(val_epoch_acc)
        
        history['train_loss'].append(epoch_loss)
        history['train_acc'].append(epoch_acc.item())
        history['val_loss'].append(val_epoch_loss)
        history['val_acc'].append(val_epoch_acc.item())
        
        print(f"Epoch {epoch+1} | Tr Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} | Val Loss: {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f} | Time: {time.time()-start_time:.0f}s")
        
        if val_epoch_acc > best_acc:
            best_acc = val_epoch_acc
            best_model_wts = copy.deepcopy(model.state_dict())

    print(f"\nBest Validation Accuracy: {best_acc:.4f}")
    model.load_state_dict(best_model_wts)
    return model, history

# ==================================================================================
# 5. EVALUATION: STANDARD & ADVANCED METRICS
# ==================================================================================
def full_evaluation_suite(model, history, val_loader, classes):
    print("\n--- 1. Standard Training Curves ---")
    epochs = range(1, len(history['train_loss']) + 1)
    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], label='Train Loss')
    plt.plot(epochs, history['val_loss'], label='Val Loss')
    plt.title('Loss Curve')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], label='Train Acc')
    plt.plot(epochs, history['val_acc'], label='Val Acc')
    plt.title('Accuracy Curve')
    plt.legend()
    plt.show()

    print("\n--- 2. Gathering Predictions for Metrics ---")
    model.eval()
    y_true, y_pred, y_probs, features = [], [], [], []
    
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc="Inference"):
            inputs = inputs.to(CONFIG['DEVICE'])
            # Extract features for t-SNE
            feat = model.avgpool(model.attention(model.features(inputs)))
            feat = torch.flatten(feat, 1)
            features.extend(feat.cpu().numpy())
            
            # Predictions
            logits = model.kan_head(model.dropout(feat))
            outputs = torch.softmax(logits, dim=1)
            _, preds = torch.max(outputs, 1)
            
            y_probs.extend(outputs.cpu().numpy())
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_probs = np.array(y_probs)
    y_true_bin = label_binarize(y_true, classes=range(len(classes)))
    features = np.array(features)

    # --- Confusion Matrix ---
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix (Counts)')
    plt.show()
    print(classification_report(y_true, y_pred, target_names=classes, digits=4))

    # --- ROC & PR Curves ---
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    colors = cycle(['blue', 'red', 'green', 'orange', 'purple'])
    for i, color in zip(range(len(classes)), colors):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_probs[:, i])
        plt.plot(fpr, tpr, color=color, lw=2, label=f'{classes[i]} (AUC={auc(fpr, tpr):.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title('ROC Curve')
    plt.legend()

    plt.subplot(1, 2, 2)
    for i, color in zip(range(len(classes)), colors):
        p, r, _ = precision_recall_curve(y_true_bin[:, i], y_probs[:, i])
        plt.plot(r, p, color=color, lw=2, label=f'{classes[i]} (AP={average_precision_score(y_true_bin[:, i], y_probs[:, i]):.2f})')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()
    
    print("\n--- 3. Advanced Analysis (Paper Extras) ---")
    
    # A. t-SNE Plot
    print("Generating t-SNE Embedding...")
    tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto')
    # Subsample if dataset is too large (>2000 points) for speed
    if len(features) > 2000:
        idx = np.random.choice(len(features), 2000, replace=False)
        tsne_emb = tsne.fit_transform(features[idx])
        tsne_labels = y_true[idx]
    else:
        tsne_emb = tsne.fit_transform(features)
        tsne_labels = y_true
        
    plt.figure(figsize=(10, 8))
    for i, cls in enumerate(classes):
        idx = np.where(tsne_labels == i)
        plt.scatter(tsne_emb[idx, 0], tsne_emb[idx, 1], label=cls, alpha=0.6)
    plt.title('t-SNE Feature Embeddings')
    plt.legend()
    plt.show()

    # B. Calibration Curve
    plt.figure(figsize=(8, 6))
    for i, color in zip(range(len(classes)), colors):
        prob_true, prob_pred = calibration_curve((y_true == i).astype(int), y_probs[:, i], n_bins=10)
        plt.plot(prob_pred, prob_true, marker='o', label=f'{classes[i]}', color=color)
    plt.plot([0, 1], [0, 1], 'k--', label='Perfect Calibration')
    plt.title('Calibration Curve (Reliability)')
    plt.legend()
    plt.show()
    
    # C. Confidence Density (Violin)
    outcomes = ["Correct" if t==p else "Wrong" for t, p in zip(y_true, y_pred)]
    max_probs = np.max(y_probs, axis=1)
    df_conf = pd.DataFrame({'Confidence': max_probs, 'Outcome': outcomes})
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df_conf, x='Outcome', y='Confidence', palette='Set2')
    plt.title('Confidence Distribution: Correct vs Wrong')
    plt.show()

    # D. Statistical Metrics Table (MCC, Specificity)
    mcc = matthews_corrcoef(y_true, y_pred)
    print(f"\nOverall MCC Score: {mcc:.4f}")
    print(f"{'Class':<15} | {'Sensitivity':<10} | {'Specificity':<10}")
    for i in range(len(classes)):
        tp = cm[i, i]
        fn = np.sum(cm[i, :]) - tp
        fp = np.sum(cm[:, i]) - tp
        tn = np.sum(cm) - (tp + fp + fn)
        spec = tn / (tn + fp) if (tn+fp) > 0 else 0
        sens = tp / (tp + fn) if (tp+fn) > 0 else 0
        print(f"{classes[i]:<15} | {sens:.4f}     | {spec:.4f}")

# ==================================================================================
# 6. XAI VISUALIZATION
# ==================================================================================
def visualize_xai_grid(model, val_loader, classes):
    print("\n--- Generating 6x6 Grad-CAM Grid ---")
    model.eval()
    inputs, labels = next(iter(val_loader))
    while len(inputs) < 12:
        i2, l2 = next(iter(val_loader))
        inputs = torch.cat((inputs, i2), 0)
        labels = torch.cat((labels, l2), 0)
    
    inputs = inputs[:12].to(CONFIG['DEVICE'])
    labels = labels[:12]
    
    target_layer = model.features[-1]
    layer_gc = LayerGradCam(model, target_layer)
    
    fig, axes = plt.subplots(6, 6, figsize=(20, 20))
    plt.subplots_adjust(wspace=0.05, hspace=0.3)
    
    for row in range(6):
        img_indices = [row*2, row*2+1]
        for idx_in_row, img_idx in enumerate(img_indices):
            col_offset = idx_in_row * 3
            input_tensor = inputs[img_idx].unsqueeze(0)
            lbl_idx = labels[img_idx].item()
            
            attr = layer_gc.attribute(input_tensor, target=lbl_idx)
            attr = torch.nn.functional.interpolate(attr, size=(224, 224), mode='bilinear').squeeze().cpu().detach().numpy()
            
            img_disp = inputs[img_idx].cpu().detach().permute(1, 2, 0).numpy()
            img_disp = img_disp * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img_disp = np.clip(img_disp, 0, 1)
            
            axes[row, col_offset].imshow(img_disp)
            axes[row, col_offset].set_title(f"{classes[lbl_idx]}", fontsize=9)
            axes[row, col_offset].axis('off')
            
            axes[row, col_offset+1].imshow(attr, cmap='jet', alpha=0.8)
            axes[row, col_offset+1].axis('off')
            
            viz.visualize_image_attr(np.expand_dims(attr, 2), img_disp, method="blended_heat_map", sign="all", 
                                     show_colorbar=False, plt_fig_axis=(fig, axes[row, col_offset+2]), use_pyplot=False)
            axes[row, col_offset+2].set_title("Overlay", fontsize=9)
    plt.show()

# ==================================================================================
# 7. MAIN EXECUTION
# ==================================================================================
if __name__ == "__main__":
    # 1. Load Data
    train_loader, val_loader, class_names = get_combined_dataloaders()
    
    # 2. Initialize Model
    model = EffiKAN_Attention_Model(num_classes=len(class_names)).to(CONFIG['DEVICE'])
    
    # 3. Train Model
    model, history = train_model(model, train_loader, val_loader, epochs=CONFIG['EPOCHS'])
    
    # 4. Full Evaluation (Standard + Advanced)
    full_evaluation_suite(model, history, val_loader, class_names)
    
    # 5. XAI Analysis
    visualize_xai_grid(model, val_loader, class_names)
```
